```python
import pandas as pd #load and store the data in the form of CSV,excel,SQL etc.
import matplotlib.pyplot as plt
```

```python
#loading the data in the variable data
data=pd.read_csv("wine.csv")
print(data)
```

```python
print(data.head())
print(data.tail())
print(data.describe())
```

```python
#Let's answer this question does higher alcohol generally mean better wine quality?
print(data[["alcohol","quality"]]) #printing first 5 rows of these 2 columns to answer this question
#clearly on seeing the raw data I cannot know it so let's plot them
plt.plot(data["alcohol"],data["quality"])
plt.xlabel("quality")
plt.ylabel("alcohol")
plt.title("Plot between alcohol(Y) and the quality(X)")
plt.show()

plt.scatter(data["alcohol"], data["quality"])
plt.xlabel("Alcohol")
plt.ylabel("Quality")
plt.title("Alcohol vs Wine Quality")
plt.show()
```

```python
#let's check out the co-relation between alcohol and quality
print(data["alcohol"].corr(data["quality"]))
```
